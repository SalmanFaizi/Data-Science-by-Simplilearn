```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4)
```

```python
model.fit(x)
```

```python
model
```

```python
x.iloc[:, 0]
```

```python

# Visualize the results
plt.scatter(x.iloc[:, 0], x.iloc[:, 1],x.iloc[:,2], c=model.labels_, cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],model.cluster_centers_[:, 2], marker='x', color='r')
plt.title('K-means clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
km = KMeans(random_state=42)
visualizer = KElbowVisualizer(km, k=(2,10))
 
visualizer.fit(x)        # Fit the data to the visualizer
visualizer.show() 
```
